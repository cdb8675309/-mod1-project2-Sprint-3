```python
#Linear regression testing
#Consider the following data, which is scattered about a line with a slope of 2 and an intercept of -5:

rng = np.random.RandomState(1)
x = 10 * rng.rand(50)
y = 2 * x - 5 + rng.randn(50)
plt.scatter(x, y);

```

```python
#from above, with linear regression

from sklearn.linear_model import LinearRegression
model = LinearRegression(fit_intercept=True)

model.fit(x[:, np.newaxis], y)

xfit = np.linspace(0, 10, 1000)
yfit = model.predict(xfit[:, np.newaxis])

plt.scatter(x, y)
plt.plot(xfit, yfit);

```

```python
#The slope and intercept of the data are contained in the model's fit parameters, which in Scikit-Learn are always marked by a trailing underscore. Here the relevant parameters are coef_ and intercept_:

print("Model slope:    ", model.coef_[0])
print("Model intercept:", model.intercept_)

```

```python
#using onehot encoding to transform the red or white into binary 0 or 1 for evaluation

from sklearn.preprocessing import OneHotEncoder

#creating instance of one-hot-encoder
encoder = OneHotEncoder(handle_unknown='ignore')

#perform one-hot encoding on 'red_or_white' column 
encoder_df = pd.DataFrame(encoder.fit_transform(combiwines[['red_or_white']]).toarray())

#merge one-hot encoded columns back with original DataFrame
rw_encd_combiwines = combiwines.join(encoder_df)

rw_encd_combiwines
```
