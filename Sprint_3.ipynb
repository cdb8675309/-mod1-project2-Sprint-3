```python
#Linear regression testing
#Consider the following data, which is scattered about a line with a slope of 2 and an intercept of -5:

rng = np.random.RandomState(1)
x = 10 * rng.rand(50)
y = 2 * x - 5 + rng.randn(50)
plt.scatter(x, y);

```

```python
#from above, with linear regression

from sklearn.linear_model import LinearRegression
model = LinearRegression(fit_intercept=True)

model.fit(x[:, np.newaxis], y)

xfit = np.linspace(0, 10, 1000)
yfit = model.predict(xfit[:, np.newaxis])

plt.scatter(x, y)
plt.plot(xfit, yfit);

```

```python
#The slope and intercept of the data are contained in the model's fit parameters, which in Scikit-Learn are always marked by a trailing underscore. Here the relevant parameters are coef_ and intercept_:

print("Model slope:    ", model.coef_[0])
print("Model intercept:", model.intercept_)

```

```python
#i want to now create a model to predict if a wine is red (or white==0) based on the characteristics of that wine, from the color column.

##First version, i will split the data, and ignore only a few categories
from sklearn.model_selection import train_test_split 
#first i need to split the data
X = rw_encd_cw.drop(columns=['color', 'red_or_white', 'quality_label', 'quality', 'qual_label_cat'])

##Is this a correct value for y?
y = rw_encd_cw['color']

# #here we split our data by passing 3 arguments to train_test_split. the input set, the output set, and a keyword argument that specifies the size of our 
# #test data set. here we set it to 20% of our data. this will return a tuple, which we need to unpack into 4 variables, Xtrain,Xtest (input sets), and ytrain,ytest (output sets)

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

```python
# data normalisation with sklearn
##Machine learning algorithms like linear regression, logistic regression, neural network, PCA (principal component analysis), 
#etc., that use gradient descent as an optimization technique require data to be scaled. 
from sklearn.preprocessing import MinMaxScaler

# fit scaler on training data
norm = MinMaxScaler().fit(X_train)

# transform training data
X_train_norm = norm.transform(X_train)

# transform testing data
X_test_norm = norm.transform(X_test)
```

```python
#checking how normalization has affected the data set
df1 = pd.DataFrame(X_train_norm)
df1.describe()
```

```python
# Using Logistic Regression
from sklearn.linear_model import LogisticRegression
# Fit the model
lr = LogisticRegression()
lr.fit(X_train_norm, y_train)
```

```python
# Predict the test set

# Now you will apply the .predict() method to make predictions on test data. These predictions are stored in 'pred_lr'.

# # Make predictions
pred_lr = lr.predict(X_test_norm)
```

```python
from sklearn.metrics import confusion_matrix
print("Confusion matrix:")
print(confusion_matrix(y_test, pred_lr))
```

```python
from sklearn.metrics import accuracy_score

print("Accuracy score:", accuracy_score(y_test, pred_lr))
```

```python
Q1 = rw_encd_cw.quantile(0.25)
Q3 = rw_encd_cw.quantile(0.75)
IQR = Q3 - Q1
print(IQR)
```

```python
#creating a new dataframe with no irrelevant nor categorical data columns

rw_encd_cw_no_cats = rw_encd_cw.drop(columns=['quality', 'quality_label', 'qual_label_cat', 'red_or_white'],  axis=1)
```

```python
rw_encd_cw_no_cats.shape
```

```python
Q1 = rw_encd_cw_no_cats.quantile(0.25)
Q3 = rw_encd_cw_no_cats.quantile(0.75)
IQR = Q3 - Q1
print(IQR)
```

```python
print(rw_encd_cw_no_cats < (Q1 - 1.5 * IQR)) |(rw_encd_cw_no_cats > (Q3 + 1.5 * IQR))
```

```python
# here i remove outliers from the dataframe made just above using the IQR method from here:
# https://www.pluralsight.com/guides/cleaning-up-data-from-outliers
df_out = rw_encd_cw_no_cats[~((rw_encd_cw_no_cats < (Q1 - 1.5 * IQR)) |(rw_encd_cw_no_cats > (Q3 + 1.5 * IQR))).any(axis=1)]
print(df_out.shape)
```

```python
df_out
```

```python
##here i create a new variable name for the scrubbed dataframe, with outliers removed via the IQR score formula (1.5x on either side)
scrubbed_rw_color_encoded = df_out
```

```python
#here i create a model that will look for red or white wine, using IQR scrubbed data set

##First version, i will split the data, and ignore only a few categories
from sklearn.model_selection import train_test_split 

#first i need to split the data
X = scrubbed_rw_color_encoded.drop(columns=['color'])

##Is this a correct value for y?
y = scrubbed_rw_color_encoded['color']

# #here we split our data by passing 3 arguments to train_test_split. the input set, the output set, and a keyword argument that specifies the size of our 
# #test data set. here we set it to 20% of our data. this will return a tuple, which we need to unpack into 4 variables, Xtrain,Xtest (input sets), and ytrain,ytest (output sets)

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

```python
# data normalisation with sklearn
##Machine learning algorithms like linear regression, logistic regression, neural network, PCA (principal component analysis), 
#etc., that use gradient descent as an optimization technique require data to be scaled. 
from sklearn.preprocessing import MinMaxScaler

# fit scaler on training data
norm = MinMaxScaler().fit(X_train)

# transform training data
X_train_norm = norm.transform(X_train)

# transform testing data
X_test_norm = norm.transform(X_test)
```

```python
#checking how normalization has affected the data set
df1 = pd.DataFrame(X_train_norm)
df1.describe()
```

```python
# Using Logistic Regression
from sklearn.linear_model import LogisticRegression
# Fit the model
lr = LogisticRegression()
lr.fit(X_train_norm, y_train)
```

```python
# Predict the test set

# Now you will apply the .predict() method to make predictions on test data. These predictions are stored in 'pred_lr'.

# # Make predictions
pred_lr = lr.predict(X_test_norm)
```

```python
from sklearn.metrics import accuracy_score

print("Accuracy score:", accuracy_score(y_test, pred_lr))
```

```python
from sklearn.metrics import confusion_matrix
print("Confusion matrix:")
print(confusion_matrix(y_test, pred_lr))
```
